```python
# select * from airports where type in ('heliport', 'balloonport')

airports.loc[ airports['type'].isin(['heliport' , 'balloonport'] ) , :]
```

```python
# select * from airports where type not in ('heliport', 'balloonport')

airports.loc[ ~airports['type'].isin(['heliport' , 'balloonport'] ) , :]
```

```python
# select iso_country, type, count(*) from airports group by iso_country, type order by iso_country, type
(airports.loc[: , ['iso_country' , 'type'] ].
 groupby(['iso_country' , 'type']).size())
```

```python
# Adding a flavour of where clause.
(airports.loc[airports['iso_country'] == 'AQ' , ['iso_country' , 'type'] ].
 groupby(['iso_country' , 'type']).size())
```

```python

# Adding a the order by clause. 
# Since we need to sort by count we need to add that to the DF. 
# Also a group by results in Multi index values and 
# hence it needs to be reset before sorting

(airports.loc[airports['iso_country'] == 'AQ' , ['iso_country' , 'type'] ].
 groupby(['iso_country' , 'type']).size().to_frame('size').reset_index().
 sort_values(['iso_country' , 'size'], ascending=[True,False]))
```

```python
# select type, count(*) from airports where iso_country = 'US' 
# group by type having count(*) > 1000 order by count(*) desc


(airports.loc[airports['iso_country'] == 'US', ['type']].
 groupby(['type']).filter(lambda g : len(g) > 1000).groupby('type').size() )
```
